```python
from sklearn import linear_model as lm
```

```python
# make classification dataset using make_classification
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=0)

batch_size = 10
#make sgd classifier
classifier = lm.SGDClassifier()

for i in range(0, len(X), batch_size):
    classifier.partial_fit(X[i:i+batch_size], y[i:i+batch_size], classes=[0,1])
    
    # make prediction
    y_pred = classifier.predict(X)
    print(y_pred)
```
